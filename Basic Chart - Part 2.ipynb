```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import squarify
import pygal
import numpy as np
import datetime
```

```python
# load data
df = pd.read_csv('unemployement-rate-1948-2010.csv')
df1 = pd.read_excel('us-postage.xlsm')
df2 = pd.read_excel('world-population.xlsm')
```

```python
df_counts = df.groupby('Period').sum().reset_index()
```

### Treemap

```python
# group the values of Period
df_counts = df.groupby('Period').sum().reset_index()
# Plot Tree Map
plt.figure(figsize=(18,8))
squarify.plot(sizes=df_counts['Value'], label=df_counts['Period'], alpha=.7)
plt.axis('off')
plt.title('US Unemployment Rate', fontsize=14)
plt.show()
```

### Area Chart

```python
plt.figure(figsize=(18,8))
plt.fill_between(df2['Year'], df2['Population'])
plt.title('World Population', fontsize=15)
plt.ylabel("Numbers in billion", fontsize=13)
plt.xlabel("Year", fontsize=13)
plt.show()
```

### Stacked Area Chart

```python
# Plotly Stacked Area Chart
fig = px.area(df, x="Year", y="Value", color="Period")
fig.update_layout(
    title="US Unemployment Rate",
    xaxis_title="Year",
    yaxis_title="Unemployment Rate",
    font=dict(  
        size=13,
        color="black"
    ))
fig.show()
```

### Step Chart

```python
plt.figure(figsize=(18,8))
plt.step(df_counts['Period'], df_counts['Value'])
plt.title('US Unemployment Rate', fontsize=15)
plt.ylabel("Rate", fontsize=13)
plt.xlabel("Month", fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.step(df['Year'], df['Value'])
plt.title('US Unemployment Rate', fontsize=15)
plt.ylabel("Rate", fontsize=13)
plt.xlabel("Years", fontsize=13)
plt.show()
```
